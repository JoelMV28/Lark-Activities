```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the data set into python

```python
data=pd.read_excel("Dataset_w6.xlsx")
data
```

# 1. Find out how many countries are there in the dataset.

```python
print("Number of countries=",(data["country"].count()))
```

# 2. Examine the data by using visualizations.

```python
# 1.Pairplot
sns.pairplot(data) 
```

```python
#2.Boxplot
plt.figure(figsize=(12,10))
sns.boxplot(x="Region",y="GDP per capita (current US$)",data=data)
plt.title(" Region Vs GDP per capita (current US$) )",fontsize=16,weight="bold")
plt.xlabel("Region",fontsize="12")
plt.ylabel("GDP per capita (current US$)",fontsize="12")
plt.xticks(rotation=90)
plt.show
```

```python
#3.Barplot
plt.figure(figsize=(10,6))
sns.barplot(x="Region",y="Population in thousands (2017)",data=data)
plt.title("Region Vs Population(2017)",fontsize=16,weight="bold")
plt.xticks(rotation=90)
plt.show()
```

```python
#4.Heatmap
slice_data=data[["Surface area (km2)","Population in thousands (2017)","GDP per capita (current US$)","International trade: Imports (million US$)"]]
sns.heatmap(slice_data.corr(),annot=True)
plt.title("Heatmap",fontsize=16,weight="bold")
plt.xticks(Rotation=90)

```

# 3. Find out how many countries with respect to the region are available

# 4. On average which region has the highest
a. GDP per capita (current US$)

b. International trade: Imports (million US$)

```python
#GDP per capita (current US$)
```

```python
#Finding Mean of GDP per capita (current US$)
avg1=data.groupby("Region")[["GDP per capita (current US$)"]].mean().round(2)
avg1
```

```python
#Selecting the largest avearage to find the region with highest GDP per capita (current US$
highest_gdp=average1.nlargest(1,["GDP per capita (current US$)"])
highest_gdp
```

```python
# International trade: Imports (million US$)
```

```python
#Finding Mean of International trade: Imports (million US$)
avg2=data.groupby("Region")[["International trade: Imports (million US$)"]].mean().round(2)
avg2
```

```python
#Selecting the largest avearage to find the region with highest International trade: Imports (million US$)
highest_iti=avg2.nlargest(1,["International trade: Imports (million US$)"])
highest_iti
```

# 5. Find out which region has more consistent with respect to GDP per capita

```python
#Finding standard deviation of GDP per capita (current US$),then finding the lowest among them to select the region with consistency
Gdp=data.groupby('Region')[["GDP per capita (current US$)"]].std()
Gdp_c=Gdp.nsmallest(1,"GDP per capita (current US$)")
Gdp_c
```
